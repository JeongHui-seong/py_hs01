```python
import pandas as pd
import folium
file_path = "./starbucks_20250411113927.csv"
data = pd.read_csv(file_path)
data = data[['이름','위도','경도','주소','타입']]
data.head()
```

```python
fixdata = data[['이름','위도','경도']]
fixdata.head()
```

```python
coordinates = fixdata[['위도','경도']].to_numpy()
place = fixdata['이름'].to_numpy()

center_lat, center_long = 35.17574571, 129.0817409

m = folium.Map(location=[center_lat, center_long], zoom_start=12)
store_type = data['타입']

for i in range(len(coordinates)):
    if store_type[i] == 'pin_generalDT':
        color = "blue"
        fcolor = "orange"
    elif store_type[i] == 'pin_general':
        color = "orange"
        fcolor = "blue"
    elif store_type[i] == 'pin_generalWT':
        color = "white"
        fcolor = "black"
    elif store_type[i] == 'pin_reserve':
        color = "black"
        fcolor = "white"
    lat, long = coordinates[i]
    folium.CircleMarker(
        location=[lat, long],
        radious = 10,
        color = color,
        fill = True,
        fill_color = fcolor,
        fill_opacity = 0.5,
        tooltip = place[i]
    ).add_to(m)
    folium.Marker(
        location = [lat + 0.001, long],
        icon= folium.DivIcon(f'<div style = "font-size: 9px; color: black; background: white; width:50px; white-space: nowrap; overflow: hidden; text-overflow: ellipsis; opacity : 0.7; border-radius: 4px">{place[i]}</div>')
    ).add_to(m)
m
```
