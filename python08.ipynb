```python
import requests as req
from bs4 import BeautifulSoup as bs
url = 'https://finance.naver.com/sise/sise_market_sum.naver'

web = req.get(url)
soup = bs(web.content, 'html.parser')
title = soup.select('.tltle')
price = soup.select( '.no+td+td')
total = soup.select( '.no+td+td+td+td+td+td')


for i, (t,p,tt) in enumerate(zip(title,price,total),1):
    print(f'no.{i} 종목명:{t.text} 현재가:{p.text} 시가총액:{tt.text}')
```

```python
import requests
from bs4 import BeautifulSoup as bs

url = "https://finance.naver.com/sise/sise_market_sum.naver"
req = requests.get(url)

def stock():
    soup = bs(req.content,'html.parser')
    title = soup.select("table.type_2 td a.tltle")
    price = soup.select("table.type_2 tr td:nth-child(3)")
    
    str = ''
    for t, p in zip(title, price):
        str += f"{t.text} : {p.text}원 \n"

    print(str)
```
